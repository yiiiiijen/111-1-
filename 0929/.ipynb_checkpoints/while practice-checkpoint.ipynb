```python
print('觸犯道路交通管理處罰條例人員名單')
A = {'name':'阿程','age':34,'罪責':{'第16條':'裝置高音量喇叭或其他產生噪音器物','第17條':'汽車不依限期參加定期檢驗或臨時檢驗',
                                 '第62條':'汽車駕駛人駕駛汽車肇事','第21條':'未領有駕駛執照駕駛小型車或機車'},
                          '罰鍰':{'第16條':1600,'第17條':1500,'第62條':2500,'第21條':10500}}

B = {'name':'阿力','age':29,'罪責':{'第12條':'汽車使用吊銷、註銷之牌照'},
                          '罰鍰':{'第12條':5000}}

C = {'name':'阿瑤','age':58,'罪責':{'第82條':'未經許可在道路擺設攤位','第56條':'在禁止臨時停車處所停車'},
                          '罰鍰':{'第82條':2000,'第56條':1000}}

def guilt(name): 
    a = len(name['罪責'])
    n = []
    for b in name['罪責']:
        gu = (b,name['罪責'][b])
        n.append(gu)
    return a, n

def money(name):
    m = 0
    n = []
    for c in name['罰鍰']:
        mo = (c,name['罰鍰'][c])
        n.append(mo)
        m += name['罰鍰'][c]
    return m, n

'''犯罪次數排名'''
gua, ga = guilt(A)
gub, gb = guilt(B)
guc, gc = guilt(C)
x = [gua,gub,guc]
x1 = enumerate(x, start = 1)
x2 = (sorted(x1, key = lambda s: s[1]))
#print(x2) #[(原始位置,value)]

'''罰錢金額排名'''
moa, ma = money(A)
mob, mb = money(B)
moc, mc = money(C)
y = [moa,mob,moc]
y1 = enumerate(y, start = 1)
y2 = (sorted(y1, key = lambda s: s[1]))

while True:
    options = input('請輸入: 1.查看阿程 2.查看阿力 3.查看阿瑤 4.犯罪數排名 5.累積罰金&罰鍰數排名 6.結束程式:')
    if options == '1':
        print('Name:',A['name'],'\nAge:',A['age']) #\n分行
        print('共觸犯(條):',gua)
        print('條例內容:')
        getit(ga)
        print('罰鍰總額(元):',moa)
        print('詳細內容:')
        getit(ma)
    elif options == '2':
        print('Name:',B['name'],'\nAge:',B['age'])
        print('共觸犯(條):',gub)
        print('條例內容:')
        getit(gb)
        print('罰鍰總額(元):',mob)
        print('詳細內容:')
        getit(mb)
    elif options == '3':
        print('Name:',C['name'],'\nAge:',C['age'])
        print('共觸犯(條):',guc)
        print('條例內容:')
        getit(gc)
        print('罰鍰總額(元):',moc)
        print('詳細內容:')
        getit(mc)
    elif options == '4':
        print('阿程為1，阿力為2，阿瑤為3')
        print('(人,犯罪次數)(少->多）')
        print(x2)
    elif options == '5':
        print('阿程為1，阿力為2，阿瑤為3')
        print('(人,金額)(少->多）')
        print(y2)
    elif options == '6':
        print('Bye bye~~~')
        break
    else:
        print('輸入錯誤，請重新輸入')
        continue
```
